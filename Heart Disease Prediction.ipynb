Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('heart-disease.csv')
```

```python
# print first 5 rows of the dataset
heart_data.head()
```

```python
# print last 5 rows of the dataset
heart_data.tail()
```

```python
# number of rows and columns in the dataset
heart_data.shape
```

```python
# getting some info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of Target Variable
heart_data['target'].value_counts()
```

1 --> Defective Heart

0 --> Healthy Heart

Splitting the Features and Target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

Building a Predictive System

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)

# change the input data to a numpy array
feature_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']

input_data_as_df = pd.DataFrame([input_data], columns=feature_names)

# reshape the numpy array as we are predicting for only on instance
#input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_as_df)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```

```python
def get_user_input():
    print("Please enter the following information:")
    age = int(input("Age: "))
    sex = int(input("Sex (1 for male, 0 for female): "))
    cp = int(input("Chest Pain Type (0-3): "))
    trestbps = int(input("Resting Blood Pressure: "))
    chol = int(input("Serum Cholesterol (mg/dl): "))
    fbs = int(input("Fasting Blood Sugar (>120 mg/dl, 1 = true, 0 = false): "))
    restecg = int(input("Resting Electrocardiographic Results (0-2): "))
    thalach = int(input("Maximum Heart Rate Achieved: "))
    exang = int(input("Exercise Induced Angina (1 = yes, 0 = no): "))
    oldpeak = float(input("Oldpeak: "))
    slope = int(input("Slope of Peak Exercise ST Segment (0-2): "))
    ca = int(input("Number of Major Vessels (0-3): "))
    thal = int(input("Thalassemia (0 = normal, 1 = fixed defect, 2 = reversable defect): "))

    return (age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal)

# Get user input
input_data = get_user_input()

input_data_as_df = pd.DataFrame([input_data], columns=feature_names)
prediction = model.predict(input_data_as_df)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
